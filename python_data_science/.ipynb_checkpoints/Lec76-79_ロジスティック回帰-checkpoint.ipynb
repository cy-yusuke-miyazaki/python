### データの準備、可視化

```python
# データロード
df = sm.datasets.fair.load_pandas().data
```

```python
df.head()
```

```python
#不倫の有無を0，1で表現する新しい列「Had_Affair」を作る
#関数を作る
def affair_check(x):
    if x !=0:
        return 1
    else:
        return 0

#applyを使って新しい列用のデータを作る
df['Had_Affair'] = df['affairs'].apply(affair_check)
```

```python
df
```

```python
#Had_affair列を使ってグループ化
df.groupby('Had_Affair').mean()
```

```python
#ヒストグラムを作ってみる
#年齢でソート
sns.countplot('age', data=df.sort_values('age'), hue='Had_Affair', palette='coolwarm')
```

```python
#結婚してからの年数でソート
sns.countplot('yrs_married', data=df.sort_values('yrs_married'), hue='Had_Affair', palette='coolwarm')
```

```python
#子供の数でソート
sns.countplot('children', data=df.sort_values('children'), hue='Had_Affair', palette='coolwarm')
```

### データの前処理

```python
#職業は数値で表現されているがカテゴリ変数なので、ダミー変数に展開する
occ_dummies = pd.get_dummies(df['occupation'])
hus_occ_dummies = pd.get_dummies(df['occupation_husb'])
```

```python
occ_dummies.head()
```

```python
#読みやすいように名前をつける
occ_dummies.columns = ['occ1','occ2','occ3','occ4','occ5','occ6']
hus_occ_dummies.columns = ['hocc1','hocc2','hocc3','hocc4','hocc5','hocc6']
```

```python
#不要になったoccupationの列と、目的変数「Had_Affair」を削除する
X = df.drop(['occupation', 'occupation_husb', 'Had_Affair'], axis=1)
```

```python
#ダミー変数のDataFrameをつなげる
dummies = pd.concat([occ_dummies, hus_occ_dummies], axis=1)
```

```python
#説明変数XのDataframe
X = pd.concat([X, dummies], axis=1) 
```

```python
X.head()
```

```python
#Yに目的変数を格納する
Y = df.Had_Affair
```

```python
Y.head()
```

### 多重共線性

ダミー変数を作ったときは特に注意。
例えばmaleが1ならfamaleは0になるなど、変数同士が高い相関を示すことになる。

```python
#ひとまずダミー変数からocc1とhocc1を取り除く
#affairｓ列も、目的変数の元になっている変数なので取り除いておく
X = X.drop('occ1', axis=1)
X = X.drop('hocc1', axis=1)

X = X.drop('affairs', axis=1)
```

```python
X.head()
```

```python
#Yを1次元のarrayにする
Y = Y.values
#Y = np.array(Y) でも同じ
```

```python
Y
```

### ロジスティック回帰

```python
#LogisticRegressionクラスのインスタンスを作る
log_model = LogisticRegression()

#データを使ってモデルを作る
log_model.fit(X, Y)

#モデルの精度を確認する
log_model.score(X, Y)
```

```python
#実際の目的変数Yの平均値を計算してみる
Y.mean()
```

全員不倫していないという当て推量だと、1 - 0.32 = 0.68 なので、それよりも上で作成したモデルのほうが精度が高いことになる

```python
#どの変数が予測に寄与しているのかを見る
#変数名と、その係数を格納するDataFrameをつくる
coeff_df = DataFrame([X.columns, log_model.coef_[0]]).T
```

```python
coeff_df
```

### 学習とテスト

データを学習用とテスト用に分けて、モデルの性能を確認してみる

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

#新しいモデルを作る
log_model2 = LogisticRegression()

#学習用のデータだけでモデルをつくる
log_model2.fit(X_train, Y_train)
```

```python
#predictで予測する
class_predict = log_model2.predict(X_test)
```

```python
#精度を計算する
print(metrics.accuracy_score(Y_test, class_predict))
```
