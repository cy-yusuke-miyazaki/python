```python
from numpy.random import randn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#正規分布にしたがう乱数100個生成
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
```

```python
plt.hist(dataset2, color='indianred')
```

```python
#縦軸を度数ではなく、標準化して確率密度変数に
plt.hist(dataset1, normed=True)
```

```python
#標準化すると、度数が違うヒストグラムを重ねて比較しやすい
#alphaを指定すると透過できる
plt.hist(dataset1, normed=True, alpha=0.5, bins=20)
plt.hist(dataset2, normed=True, alpha=0.5, bins=20, color='indianred')
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
#同時分布（結合分布）
sns.jointplot(data1, data2)
```

```python
#散布図ではなく二次元のヒストグラムに（色の濃さ＝度数）
sns.jointplot(data1, data2, kind='hex')
```
